```python
# Imported necessary packages
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Assigned the JSON to variable json
json_world = 'world_bank_projects.json'
# Made JSON a DataFrame
world_bank_df = pd.read_json(json_world)
# Shows head of DataFrame
world_bank_df.head()
```

```python
# Exercise 1
# Groups the DataFrame by Country Name and Size then sorts the values from highest to lowest
top_10_countries = world_bank_df.groupby(['countryname']).size().sort_values(ascending=False)
# Displays the Top 10 Countries with the most projects
top_10_countries.head(10)
```

```python
# Exercise 2
# Loads in the JSON like a normal JSON
json_load = json.load((open('world_bank_projects.json')))
# Normalizes it
json_norm = json_normalize(json_load, 'mjtheme_namecode', ['id'])
# Displays the Top 10 Project Themes but with missing names
json_norm.head(10)
```

```python
# Calculates total for each theme without the missing data and sorts them
top_10_themes = json_norm[json_norm.name != ''].name.value_counts().sort_values(ascending=False)
# Displays the Top 10 Project Themes
top_10_themes.head(10)
```

```python
# Exercise 3
# Save json_norm to new variable
json_norm_clean = json_norm
# Sorts with name and code
sorted_json = json_norm_clean.sort_values(['code', 'name'])
# Fills all blank names with NAN
sorted_json.name[sorted_json.name == ""] = np.nan
# Displays the Top 10 with NaNs
sorted_json.head(10)
```

```python
# Backfills the NaNs
clean_json = sorted_json.fillna(method='bfill')
# Sorts the Index
clean_json.sort_index()
# Same code from Exercise 2 but now don't need to worry about missing data
top_10_clean = clean_json.name.value_counts().sort_values(ascending=False)
# Displays the updated Top 10 Project Themes
top_10_clean.head(10)
```
